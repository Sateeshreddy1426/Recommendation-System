```python
from sklearn.preprocessing import MultiLabelBinarizer

# Split genres into lists
df['genre'] = df['genre'].apply(lambda x: x.split(', ') if isinstance(x, str) else [])

# Binarize genres
mlb = MultiLabelBinarizer()
genre_features = mlb.fit_transform(df['genre'])

print(mlb.classes_)  # Show all possible genres
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df['rating_normalized'] = scaler.fit_transform(df[['rating']])
```

```python
features = np.hstack([genre_features, df[['rating_normalized']].values])
from sklearn.metrics.pairwise import cosine_similarity
similarityk= cosine_similarity(features)
```

```python
def rec_mv(target_anime, df, similarityk, top_n=5, threshold=0.5):
    try:
        # Find the index of the target anime
        target_index = df[df['genre'] == target_anime].index[0]
    except IndexError:
        # If the anime is not found, return an empty list
        print(f"Anime '{target_anime}' not found in the dataset.")
        return []

    # Get similarity scores for the target anime
    similarity_scores = similarityk[target_index]

    # Filter and sort by similarity scores above the threshold
    filtered_indices = [
        i for i, score in enumerate(similarity_scores) 
        if score >= threshold and i != target_index
    ]
    sorted_indices = sorted(filtered_indices, key=lambda x: similarity_scores[x], reverse=True)[:top_n]

    # Return the recommended anime titles as a list
    return df.iloc[sorted_indices]['genre'].tolist()
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score
def evaluate_recommendations(similaritys, test_data, train_data, threshold=0.5):
    y_true, y_pred = [], []
    for _, row in test_data.iterrows():
        target_anime = row['genre']
        print(target_anime)
        recommendations=rec_mv(target_anime,df,similarityk,top_n=5,threshold=0.5)
        

        # Check if the target anime appears in the recommendations
        y_true.append(1 if target_anime in recommendations else 0)
        y_pred.append(1 if recommendations else 0)  # Assuming we always provide recommendations

    precision = precision_score(y_true, y_pred, zero_division=1)
    recall = recall_score(y_true, y_pred, zero_division=1)
    f1 = f1_score(y_true, y_pred, zero_division=1)

    return precision, recall, f1

# Example Evaluation
precision, recall, f1 = evaluate_recommendations(similarityk, test_data, train_data, threshold=0.5)
print(f"Precision: {precision:.2f}, Recall: {recall:.2f}, F1-Score: {f1:.2f}")
```

### Interview Questions:

```python
## Can you explain the difference between user-based and item-based collaborative filtering?
```

```python
### User-based Collaborative Filtering:
### This technique recommends items based on the similarity between users' preferences. If two users have similar tastes, the system suggests items liked by one user to the other.
### Example: If both User A and User B liked "Naruto," and User A liked "One Piece," the system recommends "One Piece" to User B.
```

```python
## Item-based Collaborative Filtering:
## This approach recommends items that are similar to what the user has interacted with. It relies on similarities between items rather than users.
### Example: If a user likes "Naruto," the system recommends "Bleach" (because it's similar to "Naruto").
```

```python
### What is collaborative filtering, and how does it work?
```

```python
### Collaborative Filtering is a technique used in recommendation systems that relies on past interactions between users and items. It assumes that users who have agreed in the past will continue to do so in the future.
### How it works:
### User-Based Filtering: Finds similar users and recommends what they liked.
### Item-Based Filtering: Finds similar items and recommends them based on a user’s previous behavior.
```
